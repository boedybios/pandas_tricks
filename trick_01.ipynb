# Trick Python Pandas
### Trick 01: Menyertakan Prefix dan Suffix pada seluruh Kolom Data Frame

## Import Modules

```python
import pandas as pd
import numpy as np
```

```python
print(pd.__version__)
print(np.__version__)
```

## Persiapan Data Frame

```python
n_rows = 10
n_cols = 7
cols = tuple('ABCDEFG')

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

## Menyertakan Prefix Kolom

```python
df.add_prefix('kolom_')
```

## Menyertakan Suffix Kolom

```python
df.add_suffix('_field')
```
